```python
# Similarly to the wind direction, the time in seconds is not a useful model input
# The weather dataset has clear daily and yearly periodicities
# Using sine and cosine functions, we can compute:
# - the time of the day
# - the time of the year

# $CHALLENGIFY_BEGIN

# Collecting the timestamps
datetime_df = pd.to_datetime(DATE, format='%Y.%m.%d %H:%M:%S')
timestamps_s = datetime_df.map(pd.Timestamp.timestamp)
timestamps_s

# 24 hours x 60 minutes/hour x 60 seconds/minute
day = 24*60*60 
# 1 year in days
year = (365.2425)*day

# Creating Empty DataFrame and Storing it in variable df
time_df = pd.DataFrame()

# Day periodicity
time_df['Day_sin'] = np.sin(timestamps_s * (2 * np.pi / day))
time_df['Day_cos'] = np.cos(timestamps_s * (2 * np.pi / day))

# Year periodicity
time_df['Year_sin'] = np.sin(timestamps_s * (2 * np.pi / year))
time_df['Year_cos'] = np.cos(timestamps_s * (2 * np.pi / year))

# $CHALLENGIFY_END
```
