```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.load('../raw_data/baseline.npz', allow_pickle=True)
```

```python
SAA=data['SAA']
SZA=data['SZA']
```

```python
mean_SAA_mar = []
mean_SZA_mar = []
mean_SAA_jun = []
mean_SZA_jun = []
mean_SAA_set = []
mean_SZA_set = []
mean_SAA_dic = []
mean_SZA_dic = []


for j in range (349,354):
    for i in range(8):
        mean_SAA_mar.append(90-(round((SAA[j,i,0,0]).mean(),2)))
        mean_SZA_mar.append((round((SZA[j,i,0,0]).mean(),2)))

for j in range (878,884):
    for i in range(8):
        mean_SAA_jun.append(90-(round((SAA[j,i,0,0]).mean(),2)))
        mean_SZA_jun.append((round((SZA[j,i,0,0]).mean(),2)))

for j in range (1407,1412):
    for i in range(8):
        mean_SAA_set.append(90-(round((SAA[j,i,0,0]).mean(),2)))
        mean_SZA_set.append((round((SZA[j,i,0,0]).mean(),2)))
        
for j in range (1801,1805):
    for i in range(8):
        mean_SAA_dic.append(90-(round((SAA[j,i,0,0]).mean(),2)))
        mean_SZA_dic.append((round((SZA[j,i,0,0]).mean(),2)))
```

```python
plt.title("Solar trajectory at 21/03/2012")
plt.ylabel("Solar Elevation")
plt.xlabel("East       ---       Solar Azimuth      ---      West")
plt.plot(mean_SZA_mar,mean_SAA_mar)
```

```python
plt.title("Solar trajectory at 21/06/2012")
plt.ylabel("Solar Elevation")
plt.xlabel("East       ---       Solar Azimuth      ---      West")
plt.plot(mean_SZA_jun,mean_SAA_jun)
```

```python
plt.title("Solar trajectory at 21/09/2012")
plt.ylabel("Solar Elevation")
plt.xlabel("East       ---       Solar Azimuth      ---      West")
plt.plot(mean_SZA_set,mean_SAA_set)
```

```python
plt.title("Solar trajectory at 21/12/2012")
plt.ylabel("Solar Elevation")
plt.xlabel("East       ---       Solar Azimuth      ---      West")
plt.plot(mean_SZA_dic,mean_SAA_dic)
```
